```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd

# Load the pre-trained model (Paraphrase-Multilingual-MiniLM-L12-v2)
model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')

# Assuming df_queries, df_documents, and df_qrels are already defined
# Generate embeddings for all queries and documents
query_embeddings = model.encode(df_queries['Text'].tolist())
document_embeddings = model.encode(df_documents['Text'].tolist())

# Compute cosine similarities between each query and all documents
similarities = cosine_similarity(query_embeddings, document_embeddings)

# Initialize lists to store precision, recall, and F1 scores for each query
precisions = []
recalls = []
f1_scores = []

# For each query, get the top 5 most similar documents and calculate precision/recall/F1
for query_index in range(len(df_queries)):
    similarities_for_query = similarities[query_index]

    # Get the top 5 most similar document indices
    top_5_indices = similarities_for_query.argsort()[-5:][::-1]  # Sorted in descending order

    # Retrieve the top 5 documents
    top_5_docs = df_documents.iloc[top_5_indices]
    top_5_similarities = similarities_for_query[top_5_indices]

    # Print the top 5 documents and their similarity scores
    print(f"\nTop 5 most similar documents for Query {query_index + 1}:")
    for rank, index in enumerate(top_5_indices, start=1):
        doc_id = df_documents.iloc[index]['ID']
        similarity_score = similarities_for_query[index]
        print(f"Rank {rank}: Doc ID: {doc_id}, Similarity: {similarity_score:.4f}")

    # Get relevant documents for the current query from df_qrels
    query_id = df_queries.iloc[query_index]['ID']
    relevant_docs = df_qrels[df_qrels['Query ID'] == query_id]['Doc ID'].tolist()

    # Calculate Precision and Recall
    retrieved_docs = top_5_docs['ID'].tolist()
    relevant_retrieved = len(set(retrieved_docs) & set(relevant_docs))

    # Precision: how many of the retrieved documents are relevant
    precision = relevant_retrieved / len(retrieved_docs) if len(retrieved_docs) > 0 else 0

    # Recall: how many relevant documents were retrieved
    recall = relevant_retrieved / len(relevant_docs) if len(relevant_docs) > 0 else 0

    # F1 Score: Harmonic mean of precision and recall
    if precision + recall > 0:
        f1_score = 2 * (precision * recall) / (precision + recall)
    else:
        f1_score = 0.0

    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1_score)

    print(f"\nPrecision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1_score:.4f}")

# Compute average precision, recall, and F1 score
average_precision = np.mean(precisions)
average_recall = np.mean(recalls)
average_f1_score = np.mean(f1_scores)

print(f"\nAverage Precision: {average_precision:.4f}")
print(f"Average Recall: {average_recall:.4f}")
print(f"Average F1 Score: {average_f1_score:.4f}")
```

```python
import numpy as np
import pandas as pd
import xgboost as xgb
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
from sklearn.metrics import precision_recall_fscore_support

# Load pre-trained model (Paraphrase-Multilingual-MiniLM-L12-v2)
model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')

# Assuming you already have dataframes: df_queries, df_documents, df_qrels
# df_queries: Query dataset with 'Text' and 'ID'
# df_documents: Document dataset with 'Text' and 'ID'
# df_qrels: Relevance dataset with 'Query ID', 'Doc ID', and 'Relevance'

# Generate embeddings for queries and documents
query_embeddings = model.encode(df_queries['Text'].tolist())
document_embeddings = model.encode(df_documents['Text'].tolist())

# Initialize lists to store feature vectors, labels, and predictions
features = []
labels = []
group_sizes = []  # This will track the number of documents per query for XGBoost LTR

# Create lists to store precision, recall, and F1 for each query
query_precisions = []
query_recalls = []
query_f1_scores = []
query_relevant_docs = []  # To store which documents are relevant for each query

# Loop through each query and its relevant documents
for query_index in range(len(df_queries)):
    query_id = df_queries.iloc[query_index]['ID']
    query_text = df_queries.iloc[query_index]['Text']
    relevant_docs = df_qrels_updated[df_qrels_updated['Query ID'] == query_id]
    relevant_doc_ids = relevant_docs['Doc ID'].tolist()

    # Store relevant docs for each query
    query_relevant_docs.append(relevant_doc_ids)

    # Get the query embedding
    query_embedding = query_embeddings[query_index]

    # Generate features for each document relevant to this query
    query_features = []
    for doc_id in relevant_doc_ids:
        # Get the document text and embedding
        doc_index = df_documents[df_documents['ID'] == doc_id].index[0]
        doc_embedding = document_embeddings[doc_index]
        doc_text = df_documents.iloc[doc_index]['Text']

        # Feature 1: Cosine similarity between query and document
        cosine_sim = cosine_similarity([query_embedding], [doc_embedding])[0][0]

        # Feature 2: Document length (normalized)
        doc_length = len(doc_text.split())  # Assuming space-separated words

        # Feature 3: Query-document term overlap (simple count)
        query_terms = set(query_text.split())
        doc_terms = set(doc_text.split())
        term_overlap = len(query_terms & doc_terms) / len(query_terms)

        # Stack features for this query-document pair
        query_features.append([cosine_sim, doc_length, term_overlap])

        # Label: Relevance from df_qrels
        label = relevant_docs[relevant_docs['Doc ID'] == doc_id]['Relevance'].values[0]
        labels.append(label)

    # Add the number of relevant documents for the current query
    group_sizes.append(len(relevant_doc_ids))

    # Add the features for all documents related to this query
    features.extend(query_features)

# Convert features and labels into numpy arrays
X = np.array(features)
y = np.array(labels)

# Convert to DMatrix format for XGBoost (required for ranking tasks)
dtrain = xgb.DMatrix(X, label=y)

# Set the group sizes for the XGBoost LTR model (to group documents by query)
dtrain.set_group(group_sizes)

# Train the XGBoost model with the ranking objective
params = {
    'objective': 'rank:pairwise',  # RankNet style pairwise loss
    'eval_metric': 'ndcg',
    'learning_rate': 0.1,
    'max_depth': 6,
    'n_estimators': 100,
    'alpha': 0.1,
    'lambda': 0.1,
    'seed': 42,
}

# Train the model
bst = xgb.train(params, dtrain, num_boost_round=50)

# Now, let's rank the documents and calculate precision, recall, and F1 for each query

# Generate predictions (score for each query-document pair)
dtest = xgb.DMatrix(X)
predictions = bst.predict(dtest)

# Rank the documents for each query based on predicted scores
for query_index in range(len(df_queries)):
    query_id = df_queries.iloc[query_index]['ID']
    relevant_docs = df_qrels_updated[df_qrels_updated['Query ID'] == query_id]
    relevant_doc_ids = relevant_docs['Doc ID'].tolist()

    # Sort the predictions for the current query and retrieve top N (e.g., top 20)
    query_predictions = predictions[query_index * len(relevant_doc_ids):(query_index + 1) * len(relevant_doc_ids)]

    # Rank documents based on prediction scores
    ranked_doc_ids = [doc_id for _, doc_id in sorted(zip(query_predictions, relevant_doc_ids), reverse=True)]

    # Apply threshold for relevance
    threshold = 0.65  # Set your threshold
    top_n = 5  # Set the top N to 20
    top_n_retrieved = [doc_id for doc_id, score in zip(ranked_doc_ids, query_predictions) if score >= threshold]

    # True positives (relevant documents in top N after applying threshold)
    true_positives = len(set(top_n_retrieved) & set(relevant_doc_ids))

    # Precision: how many of the top N retrieved documents are relevant
    precision = true_positives / len(top_n_retrieved) if len(top_n_retrieved) > 0 else 0

    # Recall: how many relevant documents were retrieved
    recall = true_positives / len(relevant_doc_ids) if len(relevant_doc_ids) > 0 else 0

    # F1-Score
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    # Store the metrics for the query
    query_precisions.append(precision)
    query_recalls.append(recall)
    query_f1_scores.append(f1)

# Output the results for each query
for query_index in range(len(df_queries)):
    print(f"\nQuery ID: {df_queries.iloc[query_index]['ID']}")
    print(f"Relevant Documents: {query_relevant_docs[query_index]}")
    print(f"Precision: {query_precisions[query_index]:.4f}")
    print(f"Recall: {query_recalls[query_index]:.4f}")
    print(f"F1-Score: {query_f1_scores[query_index]:.4f}")

# Compute average precision, recall, and F1
avg_precision = np.mean(query_precisions)
avg_recall = np.mean(query_recalls)
avg_f1 = np.mean(query_f1_scores)

print(f"\nAverage Precision: {avg_precision:.4f}")
print(f"Average Recall: {avg_recall:.4f}")
print(f"Average F1-Score: {avg_f1:.4f}")
```

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sentence_transformers import SentenceTransformer
from sklearn.metrics import precision_recall_fscore_support

# Load pre-trained SBERT model
sbert_model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')

# Example data (replace with your actual data)
# df_queries = pd.DataFrame({'ID': [1, 2], 'Text': ['query 1 text', 'query 2 text']})
# df_documents = pd.DataFrame({'ID': [1, 2, 3, 4], 'Text': ['doc 1 text', 'doc 2 text', 'doc 3 text', 'doc 4 text']})
# df_qrels = pd.DataFrame({'Query ID': [1, 1, 2], 'Doc ID': [1, 2, 3], 'Relevance': [1, 0, 1]})

# Generate embeddings for queries and documents
query_embeddings = sbert_model.encode(df_queries['Text'].tolist())
document_embeddings = sbert_model.encode(df_documents['Text'].tolist())

# Add irrelevant documents (limit number sampled)
import random

num_irrelevant_to_add = 3  # Limit irrelevant documents per query
additional_rows = []

for query_id in df_qrels['Query ID'].unique():
    relevant_docs = df_qrels[df_qrels['Query ID'] == query_id]['Doc ID'].tolist()
    all_doc_ids = df_documents['ID'].tolist()
    irrelevant_docs = list(set(all_doc_ids) - set(relevant_docs))
    sampled_irrelevant = random.sample(irrelevant_docs, min(num_irrelevant_to_add, len(irrelevant_docs)))

    for doc_id in sampled_irrelevant:
        additional_rows.append({'Query ID': query_id, 'Doc ID': doc_id, 'Relevance': 0})

# Combine with original qrels
df_qrels_updated = pd.concat([df_qrels, pd.DataFrame(additional_rows)])
print(f"Updated relevance distribution:\n{df_qrels_updated['Relevance'].value_counts()}")

# RankNet Dataset
class RankNetDataset(Dataset):
    def __init__(self, queries, documents, qrels):
        self.features = []
        self.labels = []

        # Precompute cosine similarities
        self.doc_sims = torch.tensor(np.dot(document_embeddings, document_embeddings.T))  # (D x D)

        for query_idx, query_row in queries.iterrows():
            query_id = query_row['ID']
            query_embedding = query_embeddings[query_idx]

            relevant_docs = qrels[qrels['Query ID'] == query_id]
            relevant_doc_ids = relevant_docs['Doc ID'].tolist()

            if not relevant_doc_ids:
                continue  # Skip queries without relevant docs

            # Sample irrelevant documents
            all_doc_ids = documents['ID'].tolist()
            sampled_docs = random.sample(all_doc_ids, min(50, len(all_doc_ids)))

            for doc_id_1 in relevant_doc_ids:
                for doc_id_2 in sampled_docs:
                    if doc_id_1 == doc_id_2:
                        continue  # Skip self-pairs

                    # Get relevance labels
                    rel_1 = qrels[(qrels['Query ID'] == query_id) & (qrels['Doc ID'] == doc_id_1)]['Relevance'].values[0]
                    rel_2 = qrels[(qrels['Query ID'] == query_id) & (qrels['Doc ID'] == doc_id_2)]['Relevance']
                    rel_2 = rel_2.values[0] if not rel_2.empty else 0

                    # Pair features: cosine similarity between query and document embeddings
                    sim_1 = torch.nn.functional.cosine_similarity(
                        torch.tensor(query_embedding), torch.tensor(document_embeddings[documents.index[documents['ID'] == doc_id_1][0]]), dim=0
                    ).item()
                    sim_2 = torch.nn.functional.cosine_similarity(
                        torch.tensor(query_embedding), torch.tensor(document_embeddings[documents.index[documents['ID'] == doc_id_2][0]]), dim=0
                    ).item()

                    self.features.append([sim_1, sim_2])
                    self.labels.append(1 if rel_1 > rel_2 else 0)

        print(f"Total samples in dataset: {len(self.features)}")

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return torch.tensor(self.features[idx], dtype=torch.float32), torch.tensor(self.labels[idx], dtype=torch.float32)

# Data loader
ranknet_dataset = RankNetDataset(df_queries, df_documents, df_qrels_updated)
data_loader = DataLoader(ranknet_dataset, batch_size=32, shuffle=True)

# Define RankNet model
class RankNet(nn.Module):
    def __init__(self):
        super(RankNet, self).__init__()
        self.fc = nn.Sequential(
            nn.Linear(2, 64),
            nn.ReLU(),
            nn.Linear(64, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.fc(x)

# Initialize model, loss, optimizer
ranknet_model = RankNet()
criterion = nn.BCELoss()
optimizer = torch.optim.Adam(ranknet_model.parameters(), lr=0.001)

# Training loop
epochs = 5  # Reduced for testing
for epoch in range(epochs):
    ranknet_model.train()
    epoch_loss = 0
    for batch_features, batch_labels in data_loader:
        optimizer.zero_grad()
        outputs = ranknet_model(batch_features).squeeze()
        loss = criterion(outputs, batch_labels)
        loss.backward()
        optimizer.step()
        epoch_loss += loss.item()

    print(f"Epoch {epoch + 1}/{epochs}, Loss: {epoch_loss:.4f}")
```

```python
import torch.nn.functional as F

# Evaluation
ranknet_model.eval()
all_precisions = []
all_recalls = []
all_f1_scores = []

print("\nEvaluation Metrics Per Query:")

with torch.no_grad():
    for query_index in range(len(df_queries)):
        query_id = df_queries.iloc[query_index]['ID']

        # Get the query embedding for the current query
        query_embedding = query_embeddings[query_index]

        relevant_docs = df_qrels_updated[df_qrels_updated['Query ID'] == query_id]
        relevant_doc_ids = relevant_docs['Doc ID'].tolist()

        # Calculate scores for all documents
        scores = []
        doc_ids = df_documents['ID'].tolist()
        for doc_id in doc_ids:
            doc_index = df_documents[df_documents['ID'] == doc_id].index[0]

            # Get the document embedding for the current document
            doc_embedding = document_embeddings[doc_index]

            # Use PyTorch's cosine similarity
            cosine_sim = F.cosine_similarity(
                torch.tensor(query_embedding).unsqueeze(0),
                torch.tensor(doc_embedding).unsqueeze(0),
                dim=1
            ).item()
            scores.append((cosine_sim, doc_id))

        # Rank documents by score
        ranked_docs = sorted(scores, key=lambda x: x[0], reverse=True)
        ranked_doc_ids = [doc[1] for doc in ranked_docs]

        # Calculate precision, recall, and F1-score for this query
        retrieved_relevant = [doc_id for doc_id in ranked_doc_ids if doc_id in relevant_doc_ids]
        precision = len(retrieved_relevant) / len(ranked_doc_ids) if ranked_doc_ids else 0
        recall = len(retrieved_relevant) / len(relevant_doc_ids) if relevant_doc_ids else 0
        f1 = (2 * precision * recall / (precision + recall)) if (precision + recall) > 0 else 0

        all_precisions.append(precision)
        all_recalls.append(recall)
        all_f1_scores.append(f1)

        # Display metrics for current query
        print(f"Query ID: {query_id}")
        print(f"Relevant Documents: {relevant_doc_ids}")
        print(f"Precision: {precision:.4f}")
        print(f"Recall: {recall:.4f}")
        print(f"F1-Score: {f1:.4f}\n")

# Calculate average metrics
average_precision = sum(all_precisions) / len(all_precisions) if all_precisions else 0
average_recall = sum(all_recalls) / len(all_recalls) if all_recalls else 0
average_f1 = sum(all_f1_scores) / len(all_f1_scores) if all_f1_scores else 0

print("\nAverage Evaluation Metrics:")
print(f"Average Precision: {average_precision:.4f}")
print(f"Average Recall: {average_recall:.4f}")
print(f"Average F1-Score: {average_f1:.4f}")
```

```python
import numpy as np
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics.pairwise import cosine_similarity
import lightgbm as lgb

# Load pre-trained Sentence-BERT model
sbert_model = SentenceTransformer('paraphrase-multilingual-mpnet-base-v2')

# Generate embeddings for queries and documents
query_embeddings = sbert_model.encode(df_queries['Text'].tolist())
document_embeddings = sbert_model.encode(df_documents['Text'].tolist())

# Prepare dataset for LambdaMART
def prepare_lambdamart_data(queries, documents, qrels):
    data = []

    for query_index in range(len(queries)):
        query_id = queries.iloc[query_index]['ID']
        query_embedding = query_embeddings[query_index]
        relevant_docs = qrels[qrels['Query ID'] == query_id]
        relevant_doc_ids = relevant_docs['Doc ID'].tolist()

        for doc_index in range(len(documents)):
            doc_id = documents.iloc[doc_index]['ID']
            doc_embedding = document_embeddings[doc_index]

            # Calculate cosine similarity as a feature
            cosine_sim = cosine_similarity([query_embedding], [doc_embedding])[0][0]

            # Determine relevance label
            relevance = relevant_docs[relevant_docs['Doc ID'] == doc_id]['Relevance'].values
            label = relevance[0] if len(relevance) > 0 else 0

            data.append({
                'query_id': query_id,
                'doc_id': doc_id,
                'feature': cosine_sim,
                'label': label
            })

    return pd.DataFrame(data)

# Create dataset
lambdamart_data = prepare_lambdamart_data(df_queries, df_documents, df_qrels)

# Train-test split
train_data = lambdamart_data.sample(frac=0.8, random_state=42)
test_data = lambdamart_data.drop(train_data.index)

# Prepare LightGBM datasets
train_lgb = lgb.Dataset(
    train_data[['feature']],
    label=train_data['label'],
    group=train_data.groupby('query_id').size().tolist()
)

test_lgb = lgb.Dataset(
    test_data[['feature']],
    label=test_data['label'],
    group=test_data.groupby('query_id').size().tolist(),
    reference=train_lgb
)

# LambdaMART parameters
params = {
    'objective': 'lambdarank',
    'metric': 'ndcg',
    'learning_rate': 0.05,
    'num_leaves': 31,
    'min_data_in_leaf': 20,
    'max_depth': -1,
    'verbose': -1
}

# Train LambdaMART model with early stopping
model = lgb.train(
    params,
    train_lgb,
    num_boost_round=100,
    valid_sets=[test_lgb],
    valid_names=['valid'],
    callbacks=[lgb.early_stopping(stopping_rounds=10)]
)

# Evaluate model
preds = model.predict(test_data[['feature']])
test_data['predicted_score'] = preds

# Calculate metrics
def evaluate_ranking(data):
    grouped = data.groupby('query_id')
    all_precisions, all_recalls, all_f1_scores = [], [], []

    for query_id, group in grouped:
        group = group.sort_values('predicted_score', ascending=False)
        relevant_docs = group[group['label'] > 0]['doc_id'].tolist()
        retrieved_docs = group['doc_id'].tolist()

        precision = len([doc for doc in retrieved_docs if doc in relevant_docs]) / len(retrieved_docs) if retrieved_docs else 0
        recall = len([doc for doc in retrieved_docs if doc in relevant_docs]) / len(relevant_docs) if relevant_docs else 0
        f1 = (2 * precision * recall / (precision + recall)) if (precision + recall) > 0 else 0

        all_precisions.append(precision)
        all_recalls.append(recall)
        all_f1_scores.append(f1)

    avg_precision = np.mean(all_precisions)
    avg_recall = np.mean(all_recalls)
    avg_f1 = np.mean(all_f1_scores)
    return avg_precision, avg_recall, avg_f1

average_precision, average_recall, average_f1 = evaluate_ranking(test_data)
print("\nEvaluation Metrics:")
print(f"Average Precision: {average_precision:.4f}")
print(f"Average Recall: {average_recall:.4f}")
print(f"Average F1-Score: {average_f1:.4f}")
```
